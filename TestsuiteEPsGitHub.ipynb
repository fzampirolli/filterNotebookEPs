## ğŸ“Š 3. Painel de Resultados
Veja de forma clara quais exercÃ­cios jÃ¡ estÃ£o corretos e quais precisam de atenÃ§Ã£o.

```python
import json
import os

def mostrar_status():
    if not os.path.exists('test_cache.json'):
        print("âŒ Nenhum teste executado ainda.")
        return

    with open('test_cache.json', 'r') as f:
        data = json.load(f)

    aprovados = list(data.get('approved', {}).keys())
    falhas = list(data.get('failed', {}).keys())

    print(f"âœ… APROVADOS ({len(aprovados)}):", ", ".join(sorted(aprovados)) if aprovados else "Nenhum")
    print(f"âŒ PRECISAM DE CORREÃ‡ÃƒO ({len(falhas)}):", ", ".join(sorted(falhas)) if falhas else "Nenhum")

    total = len(aprovados) + len(falhas)
    if total > 0:
        print(f"\nğŸ† Progresso: {(len(aprovados)/total)*100:.1f}%")

mostrar_status()
```

## ğŸ“‚ 4. GestÃ£o de Arquivos
Comandos Ãºteis para organizar seu trabalho.

```python
# @title ğŸ“ Criar Novo Arquivo Vazio
Nome_do_EP = "EP3_1.py" # @param {type:"string"}
Sobrescrever_se_existir = False # @param {type:"boolean"}

import os
import re

# 1. NormalizaÃ§Ã£o do nome do arquivo
Nome_do_EP = Nome_do_EP.strip()
if not Nome_do_EP.upper().startswith("EP"):
    Nome_do_EP = "EP" + Nome_do_EP

# 2. ExtraÃ§Ã£o robusta do capÃ­tulo (funciona para EP6_1 ou EP10_1)
try:
    Capitulo = re.findall(r'EP(\d+)', Nome_do_EP.upper())[0]
except IndexError:
    print("âŒ Erro: Nome do arquivo invÃ¡lido. Use o padrÃ£o EPX_Y.py")
    Capitulo = None

if Capitulo:
    caminho_pasta = f"cases_gab/cap{Capitulo}"
    caminho_completo = os.path.join(caminho_pasta, Nome_do_EP)

    # 3. Criar pasta
    if not os.path.exists(caminho_pasta):
        os.makedirs(caminho_pasta)

    # 4. LÃ³gica de criaÃ§Ã£o com alerta
    if os.path.exists(caminho_completo) and not Sobrescrever_se_existir:
        print(f"âš ï¸  PULO: O arquivo '{Nome_do_EP}' jÃ¡ existe e seu conteÃºdo foi PRESERVADO.")
        print("ğŸ’¡ Para limpÃ¡-lo, marque a opÃ§Ã£o 'Sobrescrever_se_existir'.")
    else:
        # Se Sobrescrever_se_existir for True e o arquivo existir, o 'w' limpa o conteÃºdo
        with open(caminho_completo, 'w') as f:
            f.write(f"# {Nome_do_EP}\n# Implemente sua soluÃ§Ã£o abaixo\n")

        status = "SOBREESCRITO (limpo)" if Sobrescrever_se_existir and os.path.exists(caminho_completo) else "criado"
        print(f"âœ… Arquivo {status} em: {caminho_completo}")

    print(f"\nğŸ“‚ ConteÃºdo atual de {caminho_pasta}:")
    !ls {caminho_pasta}
```

```python
# @title ğŸš€ Criar Todos os Arquivos Vazios de um CapÃ­tulo
Capitulo = 3 # @param {type:"integer"}
Total_Exercicios = 10 # @param {type:"slider", min:1, max:50, step:1}
Sobrescrever_Existentes = False # @param {type:"boolean"}

import os

caminho_pasta = f"cases_gab/cap{Capitulo}"

# 1. Criar a pasta se nÃ£o existir
if not os.path.exists(caminho_pasta):
    os.makedirs(caminho_pasta)
    print(f"ğŸ“ Pasta {caminho_pasta} criada.")

# 2. Criar os arquivos um a um com verificaÃ§Ã£o
criados = 0
pulaodos = 0

for i in range(1, Total_Exercicios + 1):
    nome_arquivo = f"EP{Capitulo}_{i}.py"
    caminho_completo = os.path.join(caminho_pasta, nome_arquivo)

    if os.path.exists(caminho_completo) and not Sobrescrever_Existentes:
        pulaodos += 1
    else:
        # Cria ou limpa o arquivo
        with open(caminho_completo, 'w') as f:
            f.write(f"# {nome_arquivo}\n# Implemente sua soluÃ§Ã£o abaixo\n")
        criados += 1

print(f"\n--- RELATÃ“RIO ---")
print(f"âœ… Novos arquivos criados: {criados}")
print(f"âš ï¸  Arquivos preservados (jÃ¡ existiam): {pulaodos}")
print(f"------------------")
print("ğŸ’¡ Atualize o menu lateral para visualizar.")
```

#### ğŸ”‘ Como gerar o seu Personal Access Token (PAT)

Desde 2021, o GitHub nÃ£o permite o uso da sua senha comum para enviar arquivos via terminal ou Colab. Ã‰ necessÃ¡rio gerar um **Token de Acesso**, que funciona como uma senha especÃ­fica para aplicaÃ§Ãµes.

##### Passo a Passo:

1. **Aceda Ã s DefiniÃ§Ãµes:** No [GitHub](https://github.com/), clique na sua foto de perfil (canto superior direito) e selecione **Settings**.
2. **Developer Settings:** No menu lateral esquerdo, role atÃ© o fim e clique em **<> Developer settings**.
3. **Tokens Classic:** Clique em **Personal access tokens** e depois em **Tokens (classic)**.
4. **Criar Novo Token:** Clique no botÃ£o **Generate new token** e escolha **Generate new token (classic)**.
5. **ConfiguraÃ§Ã£o:**
* **Note:** Escreva algo como `Colab_Lab_PI`.
* **Expiration:** Escolha um prazo (ex: 90 dias ou No expiration).
* **Scopes:** Selecione a primeira caixa: **`repo`** (isto Ã© fundamental para permitir o envio de ficheiros).


6. **Gerar e Guardar:** Role atÃ© ao fim e clique em **Generate token**.
