## 6ï¸âƒ£ RelatÃ³rio de Testes

Visualiza o relatÃ³rio JSON gerado com os resultados dos testes.

````python
import json

# LÃª e exibe o relatÃ³rio de forma formatada
try:
    with open('test_cache.json', 'r') as f:
        relatorio = json.load(f)

    print("ğŸ“Š RELATÃ“RIO DE TESTES")
    print("="*60)

    # Aprovados
    print("\nâœ… APROVADOS:")
    print("-"*60)
    if relatorio.get('approved'):
        for ep in sorted(relatorio['approved'].keys()):
            print(f"  âœ“ {ep}")
        print(f"\n  Total: {len(relatorio['approved'])} arquivo(s)")
    else:
        print("  Nenhum arquivo aprovado")

    # Falhas
    print("\nâŒ FALHAS:")
    print("-"*60)
    if relatorio.get('failed'):
        for ep in sorted(relatorio['failed'].keys()):
            print(f"  âœ— {ep}")
        print(f"\n  Total: {len(relatorio['failed'])} arquivo(s)")
    else:
        print("  Nenhuma falha! ğŸ‰")

    # EstatÃ­sticas
    total = len(relatorio.get('approved', {})) + len(relatorio.get('failed', {}))
    aprovados = len(relatorio.get('approved', {}))
    taxa = (aprovados / total * 100) if total > 0 else 0

    print("\n" + "="*60)
    print("ğŸ“ˆ ESTATÃSTICAS")
    print("="*60)
    print(f"  Total de testes: {total}")
    print(f"  Aprovados: {aprovados}")
    print(f"  Falhas: {len(relatorio.get('failed', {}))}")
    print(f"  Taxa de sucesso: {taxa:.1f}%")
    print("="*60)

except FileNotFoundError:
    print("âš ï¸ Arquivo test_cache.json nÃ£o encontrado.")
    print("Execute a cÃ©lula de testes primeiro.")
except json.JSONDecodeError:
    print("âš ï¸ Erro ao ler o arquivo test_cache.json")
    print("\nConteÃºdo bruto:")
    !cat test_cache.json
````
