## üìä 3. Painel de Resultados
Veja de forma clara quais exerc√≠cios j√° est√£o corretos e quais precisam de aten√ß√£o.

```python
import json
import os

def mostrar_status():
    if not os.path.exists('test_cache.json'):
        print("‚ùå Nenhum teste executado ainda.")
        return

    with open('test_cache.json', 'r') as f:
        data = json.load(f)

    aprovados = list(data.get('approved', {}).keys())
    falhas = list(data.get('failed', {}).keys())

    print(f"‚úÖ APROVADOS ({len(aprovados)}):", ", ".join(sorted(aprovados)) if aprovados else "Nenhum")
    print(f"‚ùå PRECISAM DE CORRE√á√ÉO ({len(falhas)}):", ", ".join(sorted(falhas)) if falhas else "Nenhum")

    total = len(aprovados) + len(falhas)
    if total > 0:
        print(f"\nüèÜ Progresso: {(len(aprovados)/total)*100:.1f}%")

mostrar_status()
```

#### üîë Como gerar o seu Personal Access Token (PAT)

Desde 2021, o GitHub n√£o permite o uso da sua senha comum para enviar arquivos via terminal ou Colab. √â necess√°rio gerar um **Token de Acesso**, que funciona como uma senha espec√≠fica para aplica√ß√µes.

##### Passo a Passo:

1. **Aceda √†s Defini√ß√µes:** No [GitHub](https://github.com/), clique na sua foto de perfil (canto superior direito) e selecione **Settings**.
2. **Developer Settings:** No menu lateral esquerdo, role at√© o fim e clique em **<> Developer settings**.
3. **Tokens Classic:** Clique em **Personal access tokens** e depois em **Tokens (classic)**.
4. **Criar Novo Token:** Clique no bot√£o **Generate new token** e escolha **Generate new token (classic)**.
5. **Configura√ß√£o:**
* **Note:** Escreva algo como `Colab_Lab_PI`.
* **Expiration:** Escolha um prazo (ex: 90 dias ou No expiration).
* **Scopes:** Selecione a primeira caixa: **`repo`** (isto √© fundamental para permitir o envio de ficheiros).


6. **Gerar e Guardar:** Role at√© ao fim e clique em **Generate token**.

> ‚ö†Ô∏è **IMPORTANTE:** Copie o c√≥digo (token) que aparecer√° no ecr√£ e guarde-o num local seguro (como um bloco de notas ou gestor de senhas). **Voc√™ n√£o conseguir√° v√™-lo novamente ap√≥s fechar a aba.**
