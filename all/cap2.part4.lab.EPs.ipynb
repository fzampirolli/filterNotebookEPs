### EP2_1 üìö Fun√ß√£o Delta ‚Äî Introdu√ß√£o a Fun√ß√µes

Esta atividade tem como objetivo ensinar o processo de submiss√£o e avalia√ß√£o de c√≥digos que utilizam **fun√ß√µes**.

* Voc√™ deve implementar um programa que calcule o discriminante (delta) de uma equa√ß√£o de segundo grau.

#### `#[py]#`
#### **üêç Python (`Q1.py`)**

````python
#[py]#
def delta(a, b, c):
    d = b * b - 4 * a * c
    return d

# Programa principal
valor = delta(5, -2, 4)
print("O delta de 5x^2 ‚Äì 2x + 4 √© %.1f" % valor)
````
