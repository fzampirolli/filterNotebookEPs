````python
# @title ğŸŒŸ SimulaÃ§Ã£o - O Teste de Mandelbrot: Fugindo do Infinito

from IPython.display import HTML

conteudo = """
<!-- SIMULAÃ‡ÃƒO INTERATIVA -->
  <hr style="margin: 30px 0; border: 0; border-top: 1px solid #eee;">
  <div
    style="padding: 20px; background: #1a1a1a; color: #eee; border-radius: 8px;">
    <h3 style="color: #fff; text-align: center; margin-top: 0;">ğŸ® Simulador
      Visual de Agrupamento</h3>
    <p style="text-align: center; color: #fbc02d; margin-bottom: 15px;">â–¶ï¸
      Clique em "Iniciar" e veja o algoritmo fazer VÃRIAS iteraÃ§Ãµes atÃ©
      convergir!<br><small>(Seu programa faz sÃ³ a primeira iteraÃ§Ã£o, mas aqui
        vocÃª vÃª o processo completo)</small></p>
    <canvas id="kmCanvas"
      style="width: 100%; background: #000; border-radius: 4px; border: 1px solid #444;"
      width="500" height="200"></canvas>
    <div id="kmLoading"
      style="display: none; text-align: center; margin: 10px; color: #fbc02d;">
      Agrupando... â³</div>
    <div
      style="display: flex; justify-content: space-around; margin: 15px 0; text-align: center; background: #2a2a2a; padding: 10px; border-radius: 5px;">
      <div>
        <div id="kmIters"
          style="font-size: 18px; font-weight: bold; color: #00d4ff;">0</div>
        <div style="font-size: 12px;">IteraÃ§Ãµes</div>
      </div>
      <div>
        <div id="kmNpts"
          style="font-size: 18px; font-weight: bold; color: #00d4ff;">150</div>
        <div style="font-size: 12px;">Pontos</div>
      </div>
      <div>
        <div id="kmKval"
          style="font-size: 18px; font-weight: bold; color: #00d4ff;">4</div>
        <div style="font-size: 12px;">Grupos (K)</div>
      </div>
    </div>
    <div style="text-align: center; margin-bottom: 15px;"><label
        style="font-size: 14px;">NÃºmero de grupos (K): <span id="kmKdisp"
          style="font-weight: bold; color: #00d4ff;">4</span></label><br><input
        id="kmSlider" style="width: 100%; margin-top: 5px;" max="8" min="2"
        type="range" value="4"></div>
    <div
      style="display: flex; gap: 8px; justify-content: center; flex-wrap: wrap;">
      <button id="kmStart"
        style="padding: 8px 16px; cursor: pointer; background: #0056b3; color: white; border: none; border-radius: 4px;">â–¶
        Iniciar</button> <button id="kmAdd"
        style="padding: 8px 16px; cursor: pointer; background: #444; color: white; border: none; border-radius: 4px;">+
        100 Pontos</button> <button id="kmReset"
        style="padding: 8px 16px; cursor: pointer; background: #666; color: white; border: none; border-radius: 4px;">ğŸ”„
        Resetar</button></div>
    <div id="kmPalette"
      style="display: flex; justify-content: center; gap: 5px; margin-top: 15px; flex-wrap: wrap;">
    </div>
  </div>
  <script>
    (function() {
      const canvas = document.getElementById('kmCanvas');
      const ctx = canvas.getContext('2d');
      let points = [];
      let centers = [];
      let assignments = [];
      let k = 4;
      let iteration = 0;
      let anim;

      const randGray = () => Math.floor(Math.random() * 256);

      const initPoints = (n = 150) => {
        points = [];
        for (let i = 0; i < n; i++) points.push(randGray());
      };

      const initCenters = () => {
        centers = [];
        const shuffled = [...points].sort(() => Math.random() - 0.5);
        for (let i = 0; i < k; i++) {
          centers.push(i < shuffled.length ? shuffled[i] : randGray());
        }
      };

      const draw = () => {
        ctx.fillStyle = '#000';
        ctx.fillRect(0, 0, canvas.width, canvas.height);

        points.forEach((p, i) => {
          const x = (p / 255) * canvas.width;
          const cl = assignments[i] !== undefined ? assignments[i] : -1;
          ctx.fillStyle = cl === -1 ? '#555' : `rgb(${p},${p},${p})`;
          ctx.fillRect(x, 0, 1.5, canvas.height);
        });

        centers.forEach(c => {
          const x = (c / 255) * canvas.width;
          ctx.strokeStyle = '#00d4ff';
          ctx.lineWidth = 3;
          ctx.strokeRect(x - 4, 10, 8, canvas.height - 20);
        });

        const pal = document.getElementById('kmPalette');
        pal.innerHTML = '';
        centers.sort((a, b) => a - b).forEach(c => {
          const div = document.createElement('div');
          div.style.cssText =
            `width:35px; height:35px; background:rgb(${c},${c},${c}); border:2px solid #00d4ff; display:flex; align-items:center; justify-content:center; font-size:9px; color:#ff4d4d; font-weight:bold;`;
          div.textContent = Math.round(c);
          pal.appendChild(div);
        });
      };

      const step = () => {
        assignments = points.map(p => {
          let minD = Infinity,
            best = 0;
          centers.forEach((c, i) => {
            const d = Math.abs(p - c);
            if (d < minD) {
              minD = d;
              best = i;
            }
          });
          return best;
        });

        const sums = Array(k).fill(0),
          counts = Array(k).fill(0);
        points.forEach((p, i) => {
          sums[assignments[i]] += p;
          counts[assignments[i]]++;
        });

        let changed = false;
        centers.forEach((c, i) => {
          if (counts[i] > 0) {
            const newVal = Math.round(sums[i] / counts[i]);
            if (Math.abs(c - newVal) > 0.5) {
              centers[i] = newVal;
              changed = true;
            }
          }
        });

        draw();
        iteration++;
        document.getElementById('kmIters').textContent = iteration;
        if (changed && iteration < 100) {
          // anim = requestAnimationFrame(step);
          anim = setTimeout(step, 1000);
        } else {
          document.getElementById('kmLoading').style.display = 'none';
        }
      };

      const reset = () => {
        cancelAnimationFrame(anim);
        document.getElementById('kmLoading').style.display = 'none';
        k = parseInt(document.getElementById('kmSlider').value);
        document.getElementById('kmKval').textContent = k;
        document.getElementById('kmKdisp').textContent = k;
        initPoints(points.length || 150);
        initCenters();
        assignments = [];
        iteration = 0;
        document.getElementById('kmIters').textContent = 0;
        document.getElementById('kmNpts').textContent = points.length;
        draw();
      };

      document.getElementById('kmStart').onclick = () => {
        document.getElementById('kmLoading').style.display = 'block';
        iteration = 0;
        step();
      };
      document.getElementById('kmReset').onclick = reset;
      document.getElementById('kmAdd').onclick = () => {
        for (let i = 0; i < 100; i++) points.push(randGray());
        document.getElementById('kmNpts').textContent = points.length;
        draw();
      };
      document.getElementById('kmSlider').oninput = e => {
        document.getElementById('kmKdisp').textContent = e.target.value;
      };

      reset();
    })();
  </script>

"""
HTML(conteudo)
````
