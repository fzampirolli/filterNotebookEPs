### EP3_1 💼 Vendedor — Comissão e Metas

Nesta atividade, o objetivo é calcular a comissão de um vendedor e verificar se ele atingiu a meta estipulada pela empresa baseada em seu salário fixo.

$$C = V \times 0.20$$

$$C \ge S \times 0.50$$

📌 **Importante**:

| 3500.00 | Nao atingiu meta de vendas |

### EP3_2 🧮 Equação do Segundo Grau — Bhaskara

Nesta atividade, você deve criar um programa que encontre as raízes de uma equação do segundo grau na forma:

$$ax^2 + bx + c = 0$$

* Leia **3 valores inteiros** correspondentes aos coeficientes $a$, $b$ e $c$.
* Calcule o discriminante ($\Delta$) utilizando a fórmula:

$$\Delta = b^2 - 4ac$$

* Avalie o valor de $\Delta$ para decidir a saída:

1.  **Se $\Delta < 0$**: A equação não possui raízes reais.
    * Imprima a mensagem: `Sem Solucao real!`
    * Na linha seguinte, imprima o valor de $\Delta$ formatado com **duas casas decimais** (ex: `Delta = -8.00`).
2.  **Se $\Delta = 0$**: A equação possui uma única raiz real.
    * Imprima o texto `x = ` seguido pelo valor da raiz formatada com **duas casas decimais**.
3.  **Se $\Delta > 0$**: A equação possui duas raízes reais distintas.
    * Imprima as duas raízes em linhas separadas.
    * Utilize os prefixos `x1 = ` para a menor raiz e `x2 = ` para a maior raiz.
    * Ambas devem estar formatadas com **duas casas decimais**.

📌 **Importante**:

* **🚫 Sem Acentos:** Utilize `Solucao` e nao *Solução* na mensagem de erro.
* Utilize a fórmula de Bhaskara para encontrar as raízes:

$$x = \frac{-b \pm \sqrt{\Delta}}{2a}$$

---

#### 💡 Dicas de Implementação

* Pode utilizar funções prontas de bibliotecas matemáticas para calcular a raiz quadrada.
* **Python**: `import math` e use `math.sqrt(x)`.
* **Java**: `import java.lang.Math;` e use `Math.sqrt(x)`.

---

#### 📌 Exemplos

| :--- | :--- |

| 4<br>-4<br>1 | x = 0.50 |
| 2<br>5<br>-4 | x1 = -3.14<br>x2 = 0.64 |

*(Nota: No último exemplo, observe que x1 é a menor raiz e x2 a maior).*

### EP3_3 🎓 Conceito Final — Notas e Classificação

Nesta atividade, você deve desenvolver um programa que calcule a Nota Final ($NF$) e o Conceito Final ($CF$) de um aluno, baseando-se em notas de testes e provas.

* Leia **3 notas de testes** ($t_1, t_2, t_3$) lidas uma a uma.
* Leia **2 notas de provas** ($P_1, P_2$) lidas uma a uma.

Para calcular a Nota Final ($NF$), primeiro calcule a **média aritmética** dos testes ($T$):

$$T = \frac{t_1 + t_2 + t_3}{3}$$

Em seguida, aplique a fórmula da média ponderada:

$$NF = 0.2 \times T + 0.4 \times P_1 + 0.4 \times P_2$$

Defina o **Conceito Final** ($CF$) de acordo com a tabela abaixo:

* **A**: se $9.0 \le NF \le 10.0$
* **B**: se $7.5 \le NF < 9.0$
* **C**: se $6.0 \le NF < 7.5$
* **D**: se $5.0 \le NF < 6.0$
* **F**: se $NF < 5.0$

**Saída:**
1.  Imprima a **Nota Final** formatada com **duas casas decimais**.
2.  Na linha seguinte, imprima o **Conceito Final** (apenas a letra).

---

#### 📌 Exemplo

| Entrada | Saída |
| :--- | :--- |
| 1<br>2<br>4<br>5<br>7 | 5.27<br>D |

*(Explicação do exemplo: Média dos testes $T = (1+2+4)/3 = 2.333...$ A Nota Final é $0.2(2.333) + 0.4(5) + 0.4(7) \approx 0.466 + 2.0 + 2.8 = 5.266...$, que arredondado fica 5.27. Como 5.27 está entre 5.0 e 6.0, o conceito é D).*

* Leia **1 número inteiro** de 6 dígitos.
* Separe o número em três partes:
    1.  **OO** (2 primeiros dígitos): Código do Planeta de Origem.
    2.  **DD** (2 dígitos do meio): Código do Planeta de Destino.
    3.  **MM** (2 últimos dígitos): Código do Modelo.

* Identifique os nomes correspondentes utilizando as tabelas abaixo:

**Tabela de Planetas (Origem e Destino):**

| Código | Planeta |
| :--- | :--- |
| 80 | Marte |
| 81 | Saturno |
| 90 | Netuno |
| 91 | HD21749b |

**Tabela de Modelos:**

| Código | Modelo |
| :--- | :--- |
| 60 | A6000 |
| 61 | B7500 |
| 62 | C9000 |

**Saída:**
* Imprima o **Nome do Planeta de Origem**.
* Imprima o **Nome do Planeta de Destino**.
* Imprima o **Nome do Modelo**.

(Cada um em uma linha separada).

---

#### 💡 Dica de Implementação

Para extrair os pares de dígitos de um número inteiro $N$ (ex: 809162), você pode utilizar operações de divisão inteira e resto (módulo):

* Para pegar os dois primeiros ($OO$): Divida por 10000.
    $$OO = N \div 10000$$
* Para pegar os dois do meio ($DD$): Pegue o resto da divisão por 10000 e divida o resultado por 100.
    $$DD = (N \pmod{10000}) \div 100$$
* Para pegar os dois últimos ($MM$): Pegue o resto da divisão por 100.
    $$MM = N \pmod{100}$$

### EP3_10 ✈️ Autonomia do Avião — Lógica e Distância

Nesta atividade, você deve determinar se um avião de carga consegue chegar ao seu destino baseando-se no peso da carga transportada e na distância a ser percorrida.

**1. Determine a Autonomia Base:**
A autonomia do avião varia de acordo com o peso da carga, conforme a tabela abaixo:

| Peso da Carga (kg) | Autonomia Base (km) |
| :--- | :--- |
| Até 50.000 | 18.000 |
| De 50.001 até 200.000 | 9.000 |
| De 200.001 até 250.000 | 3.000 |

*Nota: A capacidade máxima é 250.000 kg. Você não precisa tratar valores acima disso.*

**2. Calcule a Distância:**
O programa recebe as coordenadas de partida ($A_x, A_y$) e de destino ($B_x, B_y$). A distância ($d$) deve ser calculada usando a fórmula da distância Euclidiana:

$$d = \sqrt{(B_x - A_x)^2 + (B_y - A_y)^2}$$

**3. Verifique a Viabilidade (Saída):**
Além da autonomia base, o avião pode ter um rendimento até **10% superior** dependendo das condições de voo.

    * **TALVEZ**: Se a distância é maior que a autonomia base, mas está dentro do limite extra de 10% ($Autonomia < d \le Autonomia \times 1.10$).
    * **NAO**: Se a distância é maior que a autonomia com o adicional de 10% ($d > Autonomia \times 1.10$).

**Entrada:**
* Peso da Carga (kg)
* $A_x$ (Coordenada X de partida)
* $A_y$ (Coordenada Y de partida)

* $B_y$ (Coordenada Y de destino)

**Saída:**

* A mensagem `SIM`, `TALVEZ` ou `NAO`.

📌 **Importante**:
