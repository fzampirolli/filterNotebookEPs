### EP6_1 🔍 Busca em Matriz — Estruturas Bidimensionais

Nesta atividade, você deve implementar um programa que realize a busca de um valor específico dentro de uma matriz de números inteiros.

**Regras de Posicionamento:**
Para facilitar a leitura do usuário, a saída deve utilizar a contagem humana (iniciando em 1) e não o índice computacional (que inicia em 0).
* Se o elemento estiver no índice `[0][0]`, a posição será: **Linha 1, Coluna 1**.
* Se o elemento estiver no índice `[i][j]`, a posição será: **Linha i+1, Coluna j+1**.

**Passos do Programa:**
1.  Leia a quantidade de **linhas** ($L$) e **colunas** ($C$) da matriz.
2.  Leia cada um dos elementos para preencher a matriz $L \times C$.
3.  Leia o **número a ser buscado**.
4.  Percorra a matriz procurando o número:
    * Se encontrar, imprima a linha e a coluna (separadas por espaço).
    * Se não encontrar após percorrer toda a matriz, imprima apenas `-1`.



**Considerações:**
* É garantido que o número buscado aparece, no máximo, uma vez na matriz.
* Os elementos da matriz e o número de busca são todos inteiros.

---

#### 💡 Dicas de Leitura (Elementos na mesma linha)

Exemplo de código para ler os elementos de uma matriz no formato descrito no enunciado:

**Python:**
Use o método `split` para separar a string de entrada.
```python
n_linhas = int(input())
```

**Lógica de Preenchimento:**
* A sequência numérica começa no número **1** e aumenta de um em um até atingir o total de elementos ($L \times C$).
* **Linhas de ordem ímpar (1ª, 3ª, ...):** Devem ser preenchidas da esquerda para a direita.
* **Linhas de ordem par (2ª, 4ª, ...):** Devem ser preenchidas da direita para a esquerda.

**Passos do Programa:**
1.  Leia a quantidade de **linhas** ($L$).
2.  Leia a quantidade de **colunas** ($C$).
3.  Calcule os valores e imprima a matriz $L \times C$ no formato zig-zag.
4.  Certifique-se de que haja um caractere de **espaço** entre cada número impresso em uma mesma linha.

---

| `4`<br>`3` | `1 2 3`<br>`6 5 4`<br>`7 8 9`<br>`12 11 10` | **L1/L3:** esquerda para direita.<br>**L2/L4:** direita para esquerda. |
| `2`<br>`2` | `1 2`<br>`4 3` | Matriz $2 \times 2$ mantendo a lógica de inversão na segunda linha. |

**Considerações:**

### EP6_5 📉 Soma de Pares Abaixo da Diagonal Principal — Matrizes Quadradas

Nesta atividade, você deve implementar um programa que processe uma matriz quadrada ($n \times n$) e identifique elementos específicos para realizar uma operação de soma.

**Regras de Cálculo:**
1.  **Localização:** O programa deve observar apenas os elementos situados **abaixo da diagonal principal**.
    * A diagonal principal é composta pelos elementos onde o índice da linha é igual ao índice da coluna ($i = j$).
    * Elementos "abaixo" são aqueles onde o índice da linha é maior que o índice da coluna ($i > j$).
2.  **Condição:** Dentre esses elementos localizados abaixo da diagonal, apenas os **valores pares** devem ser somados.

**Passos do Programa:**
1.  Leia a dimensão $n$ (que define tanto o número de linhas quanto o de colunas).
2.  Leia os elementos para preencher a matriz $n \times n$.
3.  Percorra a região abaixo da diagonal principal.
4.  Identifique os números pares e calcule a sua soma total.
5.  Imprima apenas o resultado final da soma.

---

### 📌 Exemplo de Execução

Considere a matriz $3 \times 3$ abaixo:
$$
\begin{bmatrix}
6 & 3 & 2 \\
\mathbf{7} & 8 & 9 \\
\mathbf{2} & \mathbf{8} & 1
\end{bmatrix}
$$
*Elementos abaixo da diagonal principal: **7, 2 e 8**.*

| Entrada | Saída Esperada | Explicação |
| :--- | :---: | :--- |
| `3`<br>`6 3 2`<br>`7 8 9`<br>`2 8 1` | `10` | Abaixo da diagonal temos 7, 2 e 8. Os pares são 2 e 8. Soma: $2 + 8 = 10$. |
| `2`<br>`1 2`<br>`3 4` | `0` | Abaixo da diagonal temos apenas o 3 (ímpar). Soma: 0. |

---

**Considerações:**
* A diagonal principal e os elementos acima dela devem ser ignorados no cálculo.
* Se não houver números pares na região solicitada, a soma deve ser `0`.

**Regras de Transformação:**
* Considere que a dimensão **$n$** será sempre um número **par**.
* A "dobra" consiste em somar a metade inferior da matriz sobre a metade superior e, em seguida, a metade direita sobre a metade esquerda (ou vice-versa, dependendo da lógica de redução para as dimensões $n/2 \times n/2$).
* *Nota:* Com base na ilustração padrão desta atividade, cada elemento da nova matriz resultante será a soma de 4 elementos correspondentes da matriz original (um de cada quadrante).

**Passos do Programa:**
1.  Leia a dimensão $n$ da matriz quadrada.
2.  Leia os elementos para preencher a matriz original $n \times n$.
3.  Processe a lógica de dobra para reduzir a matriz para o tamanho $(n/2) \times (n/2)$.
4.  Imprima a matriz resultante, garantindo um caractere de espaço entre os números de cada linha.
