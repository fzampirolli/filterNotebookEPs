#### ⚠️ Aviso sobre a Entrada

#### 🧐 Reflexão Lógica

* **Critério de Saída Antecipada:** Se você encontrar um erro na metade do vetor, ainda é necessário verificar o restante para dar a resposta final?
* **Segurança de Índice:** Ao testar `v[i]` com `v[i+1]`, o seu loop deve terminar em `n-1` ou `n-2`?



---

#### 📌 Requisitos de Saída
O programa deve analisar os dados e imprimir rigorosamente:
1. `SIM` (caso o vetor esteja ordenado).
2. `NAO` (caso contrário - atenção à grafia sem til).

---

#### 📌 Exemplos

| Entrada | Saída Esperada |
| :--- | :--- |
| 3 <br> 1 2 3 | SIM |
| 3 <br> 3 1 2 | NAO |
| 5 <br> 10 20 20 30 40 | SIM |


#### 💡 Dicas de Leitura (Elementos na mesma linha)

Como os dados vêm na mesma linha (ex: `3 1 2`), a leitura deve ser adaptada.

Em Python, use `input().split()` para capturar a linha de números. Em Java, o `leitor.nextInt()` funcionará normalmente, pois ele ignora espaços e quebras de linha entre os números.

**Python:**
Use o método `split` para separar a string de entrada.
```python
n = int(input())
v = [0] * n

itens_linha = input().split(" ")
for i in range(n):
    v[i] = int(itens_linha[i])
```

**Java (.java):**

```java
import java.util.Scanner;

public class Main {
    public static void main(String[] args) {

      Scanner leitor = new Scanner(System.in);
      int n = leitor.nextInt();
      int[] v = new int[n];
      for (int i = 0; i < n; i++)
          v[i] = leitor.nextInt();
    }
}
```

# 🎯 Objetivo

Simular **1 passo do K-Means**:

1. Atribuir cada ponto ao centro mais próximo.
2. Recalcular a posição de cada centro (média dos pontos atribuídos).

---

# 📥 Entrada

- **Linha 1:** K números → centros iniciais  
- **Linha 2:** N números → pontos a serem agrupados  

Exemplo:

```
30 190
10 20 180 200
```

---

# 🔄 Etapa 1 — Atribuição

Para cada ponto:

```
distancia = abs(ponto - centro)
```

O ponto pertence ao centro com **menor distância**.

⚠️ Em caso de empate → escolher o **primeiro centro encontrado**.

### Exemplo

Centros: 30 e 190  

| Ponto | Centro escolhido |
|--------|------------------|
| 10     | 30 |
| 20     | 30 |
| 180    | 190 |
| 200    | 190 |

Resumo:

- Centro 1 recebeu: 10, 20  
- Centro 2 recebeu: 180, 200  

---

# 📊 Etapa 2 — Atualização dos Centros

Novo centro = média dos pontos atribuídos

```
Centro 1: (10 + 20) / 2 = 15.00
Centro 2: (180 + 200) / 2 = 190.00
```

---

# ✅ Saída Esperada

Formato obrigatório:

```
Centro X (valor_inicial): valor_novo
```

Exemplo:
